### Тестирование класса DataPreprocessor на датасете Titanic

```python
import pandas as pd

from data_preprocessor import DataPreprocessor
```

Загрузка датасета

```python
df = pd.read_csv("titanic_train_data.csv")
df.head()
```

Обработка

```python
preprocessor = DataPreprocessor(df)
result = preprocessor.fit_transform()
result.head()
```

Вывод метаданных

```python
metadata = preprocessor.metadata
print(f"Удаленные столбцы:\n{metadata.dropped_columns}\n\n"
      f"Заполнение пропусков ('столбец' : 'чем заполнялись пропуски'):\n{metadata.fill_strategy}\n\n"
      f"Исходные категориальные признаки:\n{metadata.categorical_columns}\n\n"
      f"Новые столбцы (One-Hot Encoding):\n{metadata.encoded_columns}\n\n")
```
